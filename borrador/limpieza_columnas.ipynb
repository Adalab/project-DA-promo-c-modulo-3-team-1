```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import date, datetime
import calendar
import holidays
```

```python
df = pd.read_csv('../datos/bikes.csv',index_col = 0)
```

```python
df.head()
```

```python
df['dteday']=df.dteday.apply(pd.to_datetime, format= "%d-%m-%Y")
```

```python
df.head()
```

```python
def get_season(fecha):

  Y= fecha.year
  estaciones1 = [('invierno', date(Y,  1,  1),  date(Y,  3, 20)),
                ('primavera', date(Y,  3, 21),  date(Y,  6, 20)),
                ('verano', date(Y,  6, 21),  date(Y,  9, 22)),
                ('otoño', date(Y,  9, 23),  date(Y, 12, 20)),
                ('invierno', date(Y, 12, 21),  date(Y, 12, 31))]

  for estacion, inicio, fin in estaciones1:
      if inicio <= fecha <= fin:
        return estacion        
```

```python
df['season']= df.dteday.apply(get_season)
```

```python
df.head()
```

```python
usa_feriados= holidays.US()
usa_feriados
```

```python
def detectar_feriados(fecha):
    '''Definicion: nos retorna festivos o laborales segun la fecha
       return = 1(festivo) 0 (laboral)'''

    
    if fecha in usa_feriados:
        return 1 #Festivos

    else:
        return 0 #Laboral
```

```python
df['holiday']=df['dteday'].apply(detectar_feriados)
```

```python
df.head()
```

```python
def dia_semana(col):
    return col.isoweekday()
```

```python
df['weekday']= df['dteday'].apply(dia_semana)
```

```python
df.sample(5)
```

```python
def workingday (weekday):
    '''Definicion: nos retorna dia laboral o no
       return = 1(dia laboral) 0 (fin de semana)'''


    if weekday >5 :
        return 0 #Fin de semana
    else:
        return 1 # Dia laboral
```

```python
df['workingday']= df['weekday'].apply(workingday)
```
